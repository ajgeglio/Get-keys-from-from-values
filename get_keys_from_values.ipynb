```python
# Here we have a dictionary with keys corresponding to an arbitrary list of values
Example_dict = {
    'Time': ['Time', 'T', 't', 'tim'], 
    'Torque': ['Torque', 'Tq', 'torq'], 
    'Battery': ['Battery', 'bat', 'Bat'], 
    'Current': ['Current', 'i', 'cur'], 
    'Voltage': ['Voltage', 'U', 'u', 'volt']
}
```

```python
# Its easy to list the values of a dictionary based on the keys
Example_dict['Time']
```

```python
# But what if we want to print the key for any arbitrary value that comes up?
# function if you need to get a key from a given value
def get_key(dictionary, name):
    for key, values in dictionary.items():
        for value in values:
            if name == value:
                return key
    return name

# Examples:
print(get_key(Example_dict, 't'))
print(get_key(Example_dict, 'Tq'))
print(get_key(Example_dict, 'bat'))
print(get_key(Example_dict, 'i'))
print(get_key(Example_dict, 'u'))
```

```python
#Create the expanded field names list for all possible labels
Values_list = list(Example_dict.values())
keys_list = [item for sublist in Values_list for item in sublist]
print(len(Values_list))
print(len(keys_list))
keys_list
```
